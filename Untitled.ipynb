```python
from src import *
from utils import *
import os
import os.path as osp
import datetime

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from six.moves import cPickle
import torch
```

```python
# data loader
label_df = pd.read_csv('./train_answer.csv')
n_class = 31
batch_size = 1
# 앞에서 9만번까지는 train 그 다음은 valid로 사용
train_dir_list = np.sort(glob.glob('./train/*.wav'))[:90000]
valid_dir_list = np.sort(glob.glob('./train/*.wav'))[90000:]

train_dataset = GetTrainDataset(train_dir_list, class_list=label_df.columns[1:n_class], label_image=False) 
valid_dataset = GetTrainDataset(valid_dir_list, class_list=label_df.columns[1:n_class], label_image=False) 

train_loader = DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True, num_workers=0) # 이게 worker 0이어야 에러가안남 broken pipe인지뭔지
valid_loader = DataLoader(dataset=valid_dataset, batch_size=batch_size, shuffle=True, num_workers=0) # 이게 worker 0이어야 에러가안남 broken pipe인지뭔지    

```

```python
batch_idx, (data, target) = next(enumerate(valid_loader))
mean = np.mean(data.cpu().numpy()[0, 0])
std = np.std(data.cpu().numpy()[0, 0])
print (target)
plt.imshow((data.cpu().numpy()[0, 0]-mean)/std)
plt.colorbar()
plt.show()
```
